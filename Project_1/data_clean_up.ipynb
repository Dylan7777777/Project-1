```python
#Dependencies
import pandas as pd
from pathlib import Path
```

```python
#Read the CSV file
df = pd.read_csv('realtor-data.csv')
df.head()
```

```python
#Delete extraneous column
del df['sold_date']
df.head()
```

```python
#Identify incompplete rows
df.count()
```

```python
#Drop all rows with missing information
df = df.dropna(how='any')
df.count()
```

```python
#Convert to string (object).
df.dtypes
```

```python
#Keep first duplicate data
df2 = df.drop_duplicates(keep='first')

#Delete all duplicate rows
df2 = df.drop_duplicates(keep=False)
df2.count()
```

```python
#Clean up headers
clean_realtor_data_df = df.rename(columns={"status": "STATUS",
                        "price": "PRICE", 
                        "bed": "BED", 
                        "bath": "BATH", 
                        "acre_lot": "ACRE LOT", 
                        "full_address":"FULL ADDRESS", 
                        "street": "STREET", 
                        "city": "CITY", 
                        "state": "STATE", 
                        "zip_code": "ZIP CODE", 
                        "house_size": "HOUSE SIZE"})
clean_realtor_data_df.head(20)                                
```

```python
#Export the DataFrame to a clean csv file
clean_realtor_data_df.to_csv("clean_realtor_data.cvs", index=True)
```
