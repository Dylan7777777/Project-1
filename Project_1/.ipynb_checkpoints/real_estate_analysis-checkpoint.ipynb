```python
#Create a function to fix zip_code from float and 3 digits to string 
def fix_zip(series):
    return series.astype(str).str.extract('(\d+)', expand=False).str.zfill(5)

clean_data["ZIP Code"] = fix_zip(clean_data["ZIP"])
del clean_data["ZIP"]
clean_data.head()
```

```python
#scatter plot of Size of House vs price State of New York
x_values = ny_data["HOUSE SIZE"]
y_values = ny_data["PRICE"]
(slope,intercept,rvalue,pvalue,stderr)=linregress(x_values,y_values)
regress_values = x_values * slope + intercept
line_eq = "y = "+str(round(slope,2)) + "x + "+ str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(2000,30000000),fontsize=15,color="red")
plt.title("New York House Price vs Size")
plt.xlabel("Size of House NY")
plt.ylabel("Price of House NY (in $10mln)")
plt.xlim(0,14000)
print(f"The r-squared is: {rvalue**2}")
plt.show()
```

What is more influential on the overall cost: bedrooms, size, or acres?

To determine the influence of bedrooms, size, and acres on the overall cost,  we can use a correlation matrix and a pair plot. The correlation matrix will show the relationships between these features, and the pair plot will allow us to visualize these relationships graphically.

```python
# Select the relevant columns ('BED', 'HOUSE SIZE', 'ACRE LOT', and 'PRICE') from the DataFrame
selected_columns = clean_data[['BED', 'HOUSE SIZE', 'ACRE LOT', 'PRICE']]

# Calculate the correlation matrix
correlation_matrix = selected_columns.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix: Bedrooms, Size, Acres, and Property Price')
plt.savefig("../correlation_matrix_bed_size_acres.png")
plt.show()
```

The correlation matrix is represented by a heatmap, where each cell shows the correlation coefficient between two variables. The correlation coefficient ranges from -1 to 1, where -1 indicates a perfect negative correlation, 1 indicates a perfect positive correlation, and 0 indicates no correlation. By examining the heatmap, you can see the strength and direction of the relationships between bedrooms, size, acres, and property price.

```python
# Select the relevant columns ('BED', 'HOUSE SIZE', 'ACRE LOT', and 'PRICE') from the DataFrame
selected_columns = clean_data[['BED', 'HOUSE SIZE', 'ACRE LOT', 'PRICE']]

# Sample a smaller subset of the data for visualization (adjust the sample size as needed)
sample_size = 40
random.seed(2)  # For reproducibility
sampled_data = selected_columns.sample(n=sample_size)

# Create a pair plot to visualize the relationships between these variables in the sampled data
sns.pairplot(sampled_data)
plt.suptitle('Pair Plot of Bedrooms, Size, Acres, and Property Price (Sampled Data)', y=1.02)
plt.show()
```

```python
#Insert columns for Latitude and Longitude
ma_data["Lat"]=""
ma_data["Lon"]=""
ma_data.head()
```

```python

# Import API key
#from config import geoapify_key

#Define the API parameters
params = {"apiKey":geoapify_key,"format":"json"}

#Set the base URL
base_url = "https://api.geoapify.com/v1/geocode/search?"
```

```python
# loop thgough the connecticut dataframe and search coordinates for each zip code
for index, row in ma_data.iterrows():
    
    city=row["CITY"]+", Massachusetts, United States"
    
    params["text"] = f"{city}"
    
    response=requests.get(base_url, params=params)
    
    data=response.json()
    
    
    lon = data["results"][0]["lon"]
    lat = data["results"][0]["lat"]
    
    ma_data.loc[index,"Lat"] = lat
    ma_data.loc[index,"Lon"] = lon
    
ma_data.head()
    
```

```python
#plot a scatter plot of the latitude and longitude of Pennsylvania data color the dots based on house price
plt.scatter(ma_data["Lon"], ma_data["Lat"],c=ma_data["PRICE"])
clb=plt.colorbar()
plt.title("Massacusetts Coordinates vs Price")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
clb.set_label("Price of home")
plt.savefig("../massachusetts_lat_lon_vs_price")
plt.show()
```

```python
#Potting Massachusetts houses for sale on a map diffrenciating by price

map_plot_ma = ma_data.hvplot.points("Lon","Lat",geo=True,tiles="OSM",frame_width = 725,frame_height=500,size="PRICE",scale=.01,color="CITY")
hvplot.save(map_plot_ma,"../massachusetts_map_listings_by_price.png")
map_plot_ma
```
