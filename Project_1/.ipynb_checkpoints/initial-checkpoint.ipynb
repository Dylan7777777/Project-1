```python
# Import Dependencies
from fredapi import Fred
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px

from api_keys import fred_api_key

```

# 1. Create the Fred Object

```python
fred=Fred(api_key=fred_api_key)
pd.set_option('display.max_columns',500)
```

```python
unemp_df = fred.search('Unemployment rate state',filter=('frequency','Monthly'))
unemp_df = unemp_df.query('seasonal_adjustment == "Seasonally Adjusted" and units == "Percent"')
unemp_df = unemp_df.loc[unemp_df['title'].str.contains('Unemployment Rate in')]
unemp_df = unemp_df.drop(['LASMT261982000000003','LASMT391746000000003','CMWRUR','CNEWUR','CWSTUR','CNERUR','CSOUUR'])
len(unemp_df.index)

```

```python
all_unemp =[]

for myid in unemp_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid)
    all_unemp.append(results)
    
```

```python
unemp_states=pd.concat(all_unemp,axis=1)
unemp_states
```

```python
px.line(unemp_states,title = "Unemployment Rates in US States")

```

```python
popul_df= fred.search("Resident Population in")
popul_df = popul_df.loc[popul_df["observation_start"] == '1900-01-01' ]
popul_df = popul_df.drop(['CENCPOP', 'CPACPOP', 'BEASWPOP','CMATPOP',
       'CSATLPOP', 'CWSCPOP', 'CMTNPOP', 'CWSTPOP', 'BEASEPOP', 'CESCPOP','CWNCPOP','BEAGLPOP','BEARMPOP','CMWRPOP','CSOUPOP','CNERPOP','CNEWPOP'])
popul_df.index
```

```python
all_popul =[]

for popid in popul_df.index:
    pop_results = fred.get_series(popid)
    pop_results = pop_results.to_frame(name=popid)
    all_popul.append(pop_results)
```

```python
pop_states = pd.concat(all_popul,axis=1)
pop_states.tail()
```

```python
# Calculate 10 y percentage change for each column
statepop_chg = pop_states.pct_change(10)*100
statepop_chg.tail(30)
```

```python
house_price = fred.search("All-Transactions house price index", filter=('frequency','Quarterly'))
house_price = house_price.loc[house_price["observation_start"] == '1975-01-01' ]
house_price = house_price.drop(['CESCSTHPI', 'CMATSTHPI', 'CMTNSTHPI', 'CENCSTHPI', 'CWSCSTHPI',
       'CWNCSTHPI', 'CPACSTHPI', 'CSATSTHPI','CNEWSTHPI','ATNHPIUS31084Q','USSTHPI'])
house_price.index
```

```python
all_hpi =[]

for hpi_id in house_price.index:
    hpi_results = fred.get_series(hpi_id)
    hpi_results = hpi_results.to_frame(name=hpi_id)
    all_hpi.append(hpi_results)
```

```python
hpi_states = pd.concat(all_hpi,axis=1)
hpi_states
```

```python
# Calculate 10 y percentage change for each column
statehpi_chg = hpi_states.pct_change(4)*100
statehpi_chg.tail(30)
```
